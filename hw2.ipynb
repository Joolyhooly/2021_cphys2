```python
x = np.arange(-10,10,0.01)
def f(x):
    return 0.5*x**4-3*x**3
plt.ylim(-100,100)
plt.xlim(-5,7)
plt.plot(x,f(x))
```

```python
def fp(x):
    return 2*x**3-9*x**2
```

```python
x=10
x_old =-10
eps = 0.0001
precision= 0.000001
```

```python
while (abs(x-x_old)>precision):
    x_old = x
    x=x_old - eps*fp(x_old)
```

```python
from sklearn import datasets
```

```python
diabetes =datasets.load_diabetes()
```

```python
diabetes.keys()
```

```python
print(diabetes.DESCR)
```

```python
X_age=diabetes.data[:,0]
Y=diabetes.target
```

```python
a=100;b=100

def f(a,b,X_age):
    return a*X_age+b 

def mse(a,b,X_age):
    Y_hat=np.array([f(a,b,X_age)])
    return np.mean((Y_hat-Y)**2)
```

```python
mse(a,b,X_age)
```

```python
d=0.01
n=0;
while(n<100000):
    if(mse(a+d,b,X_age)>mse(a-d,b,X_age)):
        a=a-d
    else:
        a=a+d
        
    if(mse(a,b+d,X_age)>mse(a,b-d,X_age)):
        b=b-d
    else:
        b=b+d
    n=n+1    
```

```python
a,b
```

```python
plt.plot(X_age,Y,'o')
plt.title('diabetes Linear Regression')
plt.xlabel('age')
plt.ylabel('Dia_progress')
plt.plot(X_age,f(a,b,X_age), color='red')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_age.reshape(-1,1)
X_age = X_age.reshape(-1,1)
```

```python
model=LinearRegression()
model.fit(X_age,Y)
```

```python
print('coefficiet=', model.coef_)
print('intercept=', model.intercept_)
```

```python
plt.plot(X_age,Y,'o')
plt.title('diabetes Linear Regression')
plt.xlabel('bmi')
plt.ylabel('Dia_progress')
plt.plot(X_age, model.predict(X_age),color='red')
```
